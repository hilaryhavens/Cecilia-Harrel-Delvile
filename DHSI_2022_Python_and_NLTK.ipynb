```python
def normalize_no_stops(doc):
  word_tokens_normalized = []
  for word in doc:
    word = word.lower()
    for ch in word:
      if ch in punctuation:
        word = word.strip(ch)
    if word != "" and word not in stops:
      word_tokens_normalized.append(word)
  return word_tokens_normalized
```

```python
import nltk
import gensim
import networkx as nx
from nltk.corpus import stopwords
from string import punctuation
import matplotlib.pyplot as plt
from nltk.stem.wordnet import WordNetLemmatizer
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
```

```python
nltk.download("popular")
```

```python
stop = set(stopwords.words("english"))
```

```python
for doc in nltk.corpus.gutenberg.fileids():
  print(doc)
```

```python
moby_sent_tokens = nltk.corpus.gutenberg.sents("melville-moby_dick.txt") #test paras as well
```

```python
lemma = WordNetLemmatizer()
```

```python
def normalize_no_stops_lemmatized(sent_tokens):
    """Returns normalized text, removes stops and punctuation. Lemmatized with Wordnet."""    
    normalized_sent = []
    for sent in sent_tokens:
        word_tokens_normalized = []
        for word in sent:
            word = lemma.lemmatize(word.lower())
            for ch in word:
                if ch in punctuation:
                    word = word.replace(ch, "")
            if word not in stop and word != '' and len(word) > 3:
               word_tokens_normalized.append(word)
        if word_tokens_normalized != []:
            normalized_sent.append(word_tokens_normalized)
    return normalized_sent
```

```python
help(normalize_no_stops_lemmatized)
```

```python
cleaned_doc = normalize_no_stops_lemmatized(moby_sent_tokens)
```

```python
cleaned_doc[:20]
```

```python
dictionary = gensim.corpora.Dictionary(cleaned_doc)
```

```python
doc_term_matrix = [dictionary.doc2bow(doc) for doc in cleaned_doc]
```

```python
lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = lda(doc_term_matrix, num_topics=10, id2word=dictionary, passes=50)
```

```python
topics = ldamodel.print_topics(num_topics=10, num_words=10)
```

```python
from pprint import pprint
pprint(topics)
```

```python
topic_list = []
for score, topic in topics:
  topic_vals = [tuple(top.split("*")) for top in topic.split(" +")]
  topic_list.append(topic_vals)
```

```python
topic_list
```

```python
G = nx.Graph()
```

```python
top_num = 0
for topics in topic_list:
    top_num += 1    
    G.add_node("Topic" + str(top_num), weight=0)
    for topic in topics:
        G.add_node(topic[1], weight=topic[0], size=topic[0])
        G.add_edge("Topic" + str(top_num), topic[1], weight=float(topic[0]))
```

```python
G.nodes.data()
```

```python
plt.figure(figsize=(20,10))
plt.suptitle("LDA Topic Model of Moby Dick by Herman Melville", fontsize=16)
nx.draw(G, with_labels=True)
plt.savefig("/content/moby_topic.png")
```
