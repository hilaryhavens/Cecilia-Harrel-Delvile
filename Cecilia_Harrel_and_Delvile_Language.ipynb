```python
stops = set(stopwords.words("english"))
```

Remove stopwords from language for both characters.

```python
def normalize_no_stops(doc):
  word_tokens_normalized = []
  for word in doc:
    word = word.lower()
    if word != "" and word not in stops:
      word_tokens_normalized.append(word)
  return word_tokens_normalized
```

```python
del_word_tokens_normalized = normalize_no_stops(del_word_tokens)
```

```python
har_word_tokens_normalized = normalize_no_stops(har_word_tokens)
```

Make word clouds of the most frequently used words (minus stopwords) for both characters.

```python
# This code was adapted from Stack Overflow (https://stackoverflow.com/questions/16645799/how-to-create-a-word-cloud-from-a-corpus-in-python)

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

del_word_tokens_normalized_string = ' '.join(del_word_tokens_normalized)  

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(del_word_tokens_normalized_string)
```

Import collocation tools to find collocations within text, filtering out for short characters or stop words.

```python
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
```

```python
def filter_stops(w):
  filter_stops = ""
  if len(w) < 3 or w in stops:
    filter_stops = w
  return filter_stops
```

```python
bcf = BigramCollocationFinder.from_words(del_word_tokens)
bcf.nbest(BigramAssocMeasures.likelihood_ratio, 10)
# Example of why you need to filter characters
```

```python
bcfh = BigramCollocationFinder.from_words(har_word_tokens)
bcfh.nbest(BigramAssocMeasures.likelihood_ratio, 10)
# Example of why you need to filter characters
```

Why it is necessary to take out stopwords when looking for bigrams

```python
bcf.apply_word_filter(filter_stops)
del_bigrams = bcf.nbest(BigramAssocMeasures.likelihood_ratio, 20)
pprint(del_bigrams)
```

```python
bcfh.apply_word_filter(filter_stops)
har_bigrams = bcfh.nbest(BigramAssocMeasures.likelihood_ratio, 20)
pprint(har_bigrams)
```
